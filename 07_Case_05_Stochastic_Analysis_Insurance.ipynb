## Introduction

In order to provide an overall insight into advanced simulation techniques, a few examples will be presented in this topic, representing varous types of application. The examples are constructed very simply in order to highlight the fundamental features. This also offers the opportunity of introducing some useful terms and concepts. This notebook is based on Daykin, C. D., et al. (1995). Practical Risk Theory for Actuaries. UK: Chapman & Hall.

Premiums are assumed to flow continously into a risk reserve $ U(t) $, frow which the claims $ X(t) $ are paid out. Claim payments delays are ignored. In the classical risk theory the premium income in time period $ (0,t] $ is defined to be $ P(t) = (1 + \lambda) \times P \times t $, where $ P=E[X(1)] $ is the pure risk premium per unit time and $ \lambda $ is a safety loading coefficient. No allowance is made in $ P(t) $ for expenses. We take the simple underwriting equation:


$$ U(t) - U_0 = (1+\lambda) \times P \times t - X(t) $$


where $ U_0 $ is the initial reserve, $ U(t) $ is the risk reserve at time point $ t $ and $ U(t) - U_0 $ is the net gain arising during the time period $ (0,t] $.

For this purpose it is transformed into the algorith

$$ U(t) = (1+j(t)) \times U(t-1) + (1+\lambda(t)) \times P(t) - X(t) $$

The profits are accumulated in the risk reserve and the losses are deducted from it. The coefficient $ j(t) $ is the rate of interest earned on the reserve, $ P(t) $ is the pure risk premium, $ \lambda(t) $ a safety loading and $ X(t) $ the claims. To perform the simulation the quantities $ U(0), j(t), \lambda(t), P(t) $ need to be specified, together with the parameters required for the simulation of $ X(t) $, such as $ n $, the characteristics of the claim size distribution and the mixing variable. 

First, for $ t=1 $, the claim amount is simulated using the one of the generators for a compound distribution. The outcome $ X(1) $ is substituted in the $ U(t) $ expression and $ U(1) $ is obtained. Then the procedure is repeated for $ t=2,3,\dots,T $ arriving at the sequence:

$$ U(0),U(1),U(2),\dots,U(T) $$

Conventional analytical methods are not adequate for long-term considerations. This suggests that simulation should be the principal working tool. We now show how the insurance process:

$$ U(t) = U(t-1) + B(t) + J(t) - X(t) - E(t) $$

can be simulated. Remember that $ X(t) $ means claims, $ B(t) $ means premiums, $ E(t) $ means expenses and $ J(t) $ means return on investment. The presentation is an extension of the basic model:

$$ U(t) = (1+j(t)) \times U(t-1) + (1+\lambda(t)) \times P(t) - X(t) $$

with the return on investments, decomposition into lines of business and some other features now also included.

## Basic equation

The simulation proceeds step by step from the initial accounting year $ t_0 $ to the succeeding year $ t_0 +1, t_0 +2, \dots  $. The transition from one year to the next will be carried out using a transition algorithm which treats the cash flow equation:

$$ A\left(t\right)=A\left(t-1\right)+B^\prime\left(t\right)+J\left(t\right)+X_{re}\left(t\right)+U_{new}\left(t\right)+W_{new}\left(t\right)-X^\prime\left(t\right)-E\left(t\right)-B_{re}\left(t\right)-D(t) $$

and the accounting equation

$$ U(t) = U(t-1) + B(t) + J(t) + U_{new}(t) - \Delta L_0^{'}(t) - X(t) - E(t) - R(t) - D(t) $$

in parallel. Remember that $ \Delta L_0^{'}(t) $ is any increase in net liabilities other than technical reserves, excluding any increase in borrowing or debt capital or items which are already included in other entries of the aforementioned equation. $ R(t) $ is the balance of dealings with the reinsurers. This term is not required if the accounting is net or reinsurance.

The former results in a sequence of assets amounts $ A(t_0), A(t_0+1), \dots $ which are needed for the simulation of asset movements, as well as for various cash flow considerations, in particular, when the break-up principle is applied. The accounting algorithm gives the corresponding sequence for the solvency margin $ U $ and for other variables which are required in conventional income and balance sheet statements and for many going-concern considerations. These two equations will now be employed in a simplified and modified form as follows

$$ A\left(t\right)=A\left(t-1\right)+B^\prime\left(t\right)+J^\prime\left(t\right)+\Delta^{*}A(t)-X^\prime\left(t\right)-E\left(t\right) $$

and

$$ U\left(t\right)=U\left(t-1\right)+B\left(t\right)+J\left(t\right)+\Delta^{*}A(t)-X\left(t\right)-E\left(t\right) $$

where the primed quantities indicate the amounts paid in year $ t $, as distinct from the unprimed quantities, which represent the earned or incurred amounts. The links between them are:
* $ B(t) = B^{\prime}(t) - V(t) + V(t-1) $ where $ V(t) $ is the unearned premium reserve at the end of the accounting period (year) $ t $. Earned premiums for an accounting year are those parts of premiums written in the year, or in previous years, which relate to risks borne in that accounting year. In so far as premiums written during the accounting year provide cover for risks in the next or subsequent accounting years, the part of the premium relating to those later periods is carried forward by establishing a provision for unearned premiums.
* $ X(t) = X'(t) + C(t) - C(t-1) $
